```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import gensim.downloader as api
from gensim.models import Word2Vec
```

Load Data and Train Word2Vec Model

```python
corpus=api.load("text8")
model=Word2Vec(corpus)
```

Select Words and Get Their Embeddings

```python
words = ['cat', 'dog', 'elephant', 'lion', 'bird', 'rat', 'wolf', 'cow',
         'goat', 'snake', 'rabbit', 'human', 'parrot', 'fox', 'peacock',
         'lotus', 'roses', 'marigold', 'jasmine', 'computer', 'robot',
         'software', 'vocabulary', 'machine', 'eye', 'vision',
         'grammar', 'words', 'sentences', 'language', 'verbs', 'noun',
         'transformer', 'embedding', 'neural', 'network', 'optimization']
words=[x for x in words if x in model.wv.key_to_index]
word_embedding=[model.wv[x] for x in words]
embeddings=np.array(word_embedding)
```

 Reduce Dimensionality with t-SNE

```python
tsne=TSNE(n_components=2,perplexity=2)
embeddings_2d=tsne.fit_transform(embeddings)
```

 Plot Embedding:Displays a scatter plot of the words in 2D space, labels each point with its word and displays the plot.

```python
plt.figure(figsize=(10,10))
for i,word in enumerate(words):
  plt.text(embeddings_2d[i,0],embeddings_2d[i,1],word,fontsize=10)
plt.show()
```

```python
from sentence_transformers import SentenceTransformer, util

# Load a pre-trained Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

query = "What are the advancements in NLP?"
documents = [
    "Machine learning enables advancements in NLP.",
    "Climate change is a pressing issue globally.",
    "Natural language processing allows machines to understand text."
]

# Encode query and documents
query_embedding = model.encode(query)
doc_embeddings = model.encode(documents)

# Find the most similar document to the query
results = util.semantic_search(query_embedding, doc_embeddings, top_k=1)
most_similar_doc = documents[results[0][0]['corpus_id']]
print(f"Most similar document to the query: \"{most_similar_doc}\"")
```
